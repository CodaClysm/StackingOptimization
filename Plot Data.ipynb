```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getData(fileString):
    f = open(fileString, 'r')
    generation = []
    percent = []
    raw = []
    for x in f:
        if 'Generation' in x:
            for t in x.split():
                if '%' in t:
                    t = t[:len(t)-1]
                try:
                    raw.append(float(t))
                except ValueError:
                    pass
    for i in range(len(raw)):
        if i % 2 == 0:
            generation.append(raw[i])
        else:
            percent.append(raw[i])
    generation = generation[:len(generation)-1]
    return percent
```
