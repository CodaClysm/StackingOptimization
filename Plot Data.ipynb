```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getData(fileString):
    f = open(fileString, 'r')
    generation = []
    percent = []
    raw = []
    for x in f:
        if 'Generation' in x:
            for t in x.split():
                if '%' in t:
                    t = t[:len(t)-1]
                try:
                    raw.append(float(t))
                except ValueError:
                    pass
    for i in range(len(raw)):
        if i % 2 == 0:
            generation.append(raw[i])
        else:
            percent.append(raw[i])
    generation = generation[:len(generation)-1]
    return percent
```

# Shape set 1

```python
utilizations = [];
utilizations.append(getData("datalogs/size1-set1-shape1.txt"))
utilizations.append(getData("datalogs/size1-set2-shape1.txt"))
utilizations.append(getData("datalogs/size1-set3-shape1.txt"))
utilizations.append(getData("datalogs/size1-set4-shape1.txt"))
utilizations.append(getData("datalogs/size1-set5-shape1.txt"))
utilizations.append(getData("datalogs/size1-set6-shape1.txt"))



fig = plt.figure(figsize=(15,12))

plt.plot(np.arange(len(utilizations[0])), utilizations[0], 'r-')
#plt.plot(np.arange(len(utilizations[1])), utilizations[1], 'b-')
#plt.plot(np.arange(len(utilizations[2])), utilizations[2], 'g-')
plt.plot(np.arange(len(utilizations[3])), utilizations[3], 'c-')
#plt.plot(np.arange(len(utilizations[4])), utilizations[4], 'y-')
#plt.plot(np.arange(len(utilizations[5])), utilizations[5], 'm-')


```
