```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getData(fileString):
    f = open(fileString, 'r')
    generation = []
    percent = []
    raw = []
    for x in f:
        if 'Generation' in x:
            for t in x.split():
                if '%' in t:
                    t = t[:len(t)-1]
                try:
                    raw.append(float(t))
                except ValueError:
                    pass
    for i in range(len(raw)):
        if i % 2 == 0:
            generation.append(raw[i])
        else:
            percent.append(raw[i])
    generation = generation[:len(generation)-1]
    return percent
```

# without seed

```python
utilizations = [];
utilizations.append(getData("SavedData/preseed/size1-shape1-random.txt")) #0
utilizations.append(getData("SavedData/preseed/size1-shape1-set1.txt")) #1
utilizations.append(getData("SavedData/preseed/size1-shape1-set2.txt")) #2
utilizations.append(getData("SavedData/preseed/size1-shape1-set3.txt")) #3
utilizations.append(getData("SavedData/preseed/size1-shape1-set4.txt")) #4
#utilizations.append(getData("SavedData/preseed/size1-shape1-set5.txt")) #5
#utilizations.append(getData("SavedData/preseed/size1-shape1-set6.txt")) #6



fig = plt.figure(figsize=(15,12))

plt.plot(np.arange(len(utilizations[0])), utilizations[0], 'r-')
#plt.plot(np.arange(len(utilizations[1])), utilizations[1], 'b-')
#plt.plot(np.arange(len(utilizations[2])), utilizations[2], 'g-')
#plt.plot(np.arange(len(utilizations[3])), utilizations[3], 'c-')
plt.plot(np.arange(len(utilizations[4])), utilizations[4], 'k-')
#plt.plot(np.arange(len(utilizations[5])), utilizations[5], 'y-')
#plt.plot(np.arange(len(utilizations[6])), utilizations[6], 'g-')
```
